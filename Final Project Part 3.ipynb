## Details on movies produced by different studios

The visualization below is an interactive bar chart which displays information sorted on the basis of studio. The plot shows the imdb rating, length and average worldwide gross which is the amount of money that particular studio made. This chart helps us understand about every studio and what kind of movies they make. Also, the most profitable studio out of all. The average of these yaxis columns are displayed. You can select from  imdb rating, length and worldwide gross.

```python
from ipywidgets import interactive

def bar_plot(yaxis, colors):
    y = blockbusters.groupby('studio')[yaxis].mean()
    plt.figure(figsize=(12, 8))
    ax = y.plot(kind="bar", color= colors)
    ax.set_title("Studio Vs "+ yaxis)
    ax.set_xlabel("Studio")
    ax.set_ylabel(yaxis)
    

plot = interactive(bar_plot, 
                   yaxis = ipywidgets.Dropdown(
                      value='imdb_rating',
                      options=['imdb_rating', 'length', 'worldwide_gross'],
                      description='y'
                  ),
                   colors = ipywidgets.Dropdown(
                      value='red',
                      options=['red','green','blue'],
                      description='color'
                  )
                  )
plot
```

From this visualization we see that, Pixar has made most movies with the highest IMDB rating. Whereas, Warner Bros and Lionsgate Films have movies with longest length which is closely followed by Newmarket Films. Also, Pixar and Newmarket have the highest grossing films. These information combined with other factors can tell us a lot about what kind of movies are watched the most and are the most profitable. 

## Most profitable studio over the years 

This visualization is a line chart displaying the gross income year wise, to observe which studio has been in profit or loss over the years. This helps us see the trend over the years with respect to studio. A line chart is the best way to display trends over years. This visualization allows you to select between studios and observe trends for each of them.  

```python
from ipywidgets import interactive

def bar_plot(studios):
    studio = blockbusters[blockbusters['studio']==studios]
    studio.plot('year', 'worldwide_gross')

plot = interactive(bar_plot, 
                   studios = ipywidgets.Dropdown(
                      value='Walt Disney Pictures',
                      options=['Walt Disney Pictures', 'Pixar', 'Universal Pictures',
       '20th Century Fox', 'Paramount Pictures', 'Columbia Pictures',
       'Warner Bros', 'Sony Pictures', 'Lionsgate Films',
       'Metro-Goldwyn-Mayer', 'DreamWorks'],
                      description='Studios'
                  )
                  )
plot
```

By interacting with this graph, you can gather various insights on how a studio was performing for a particular year. For example, Sony Pictures had a peak in gross after right after 1990 and big dip from 2000 to 2015, where the curve started rising up again. For Metro-Goldwyn-Mayer, there were some highs and lows between year 1975 to 1985 but they have had a constant high after 2010 and the overall graph was upward. More information can be gathered about different studios on interacting with this chart. 

## Average IMDB rating of different Genres

This bar chat displaying Genre on the x-axis and the average IMDB rating of the respective genre on the y-axis. This plot helps us see which genre is usually more popularly and highly rated. A graph like this can help a studio in deciding which genre is a better investment. 

```python
avg_rating = blockbusters.groupby('Main_Genre')['imdb_rating'].mean()
```

```python
avg_rating.plot.bar()
```

From this graph we can observe that War is the most highly rated with history following up on the second place. There is a high demand of movies based on Wars and history. We can also see that Drama is high in rating. The movies having the main genre as Family are probably least in IMDB rating.

## Highest ranked movies in 2018 and their worldwide gross

In the final visualization I wanted to display the most recent highest ranked movies available in the dataset. On the x-axis we can see the titles of the movies displayed left to right from highest to lowest rank for the year 2018. And on the y-axis we can see the worldwide gross profit made by the movie.

```python
blockbusters_year = blockbusters[blockbusters['year']==2018]
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(blockbusters_year['title'], blockbusters_year['worldwide_gross'])
plt.xticks(rotation=90)
plt.show()
```

Using visualizations like these we can sort for the movies which are most liked and profitable. Similarly, using this dataset we can find top 10 movies for other years too. From the above visualization we can see 'Black Panther' was 1st place followed by 'Avengers: Infinity War' at 2nd place both made close to 7 Million dollars. For 2018, the last movie at rank 10 is 'A Star is Born' and it made a gross of 2 Million dollars.
