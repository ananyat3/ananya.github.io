#### Creating a pivot table to group Agency Name and County together and finding the mean of the square footage of the grouped variables

```python
mean_sq_ft = pd.pivot_table(buildings, index = 'Agency Name', columns= 'County', values = 'Square Footage', aggfunc = 'mean', fill_value=0)
```

#### Writing a function for creating barplot with respect to the values that will allow us to select values on the gridmap.

```python
# BarPlot
# 1. scales
x_scs = bqplot.OrdinalScale()
y_scs = bqplot.LinearScale()

# 3. axis
x_axs = bqplot.Axis(label='Year', scale=x_scs)
y_axs = bqplot.Axis(label='Square Footage', scale=y_scs, 
                   orientation = 'vertical')

# 2. Mark!
bar_plot = bqplot.pyplot.bar(x = [], 
                         y = [],
                         scales={'x':x_scs, 'y':y_scs})

# 3. make a figure!
fig_dur = bqplot.Figure(marks=[bar_plot], axes=[x_axs,y_axs], fig_margin={'top':30, 'bottom':60, 'left':100, 'right':100})
#fig_dur
```

#### The x-scale as Ordinal as it creates a mapping from a discrete set of values to a numerical range. This is required because we will be using 'Year Acquired' as our x-axis scale for barplot. Similarly, y-scale as Linear as it creates a mapping from numerical domain to a numerical range. Plotting values of sum of square footages on y-axis is a numerical domain.

```python
# 2. scales
col_sc = bqplot.ColorScale(scheme='Red', min=np.nanmin(mean_sq_ft.values), max=np.nanmax(mean_sq_ft.values))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale() 

# 3. axis
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale=x_sc, label = 'County')
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical', label = 'Agency Name')

# 4. Marks
heat_map = bqplot.GridHeatMap(color=mean_sq_ft.values, 
                             row = mean_sq_ft.index, column = mean_sq_ft.columns, 
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                             interactions={'click':'select'}, 
                             anchor_style={'fill':'blue'})

def barplot_fig2(x, y):
    df2 = buildings[(buildings['County'] == y) & (buildings['Agency Name'] == x)]
    df2 = df2[df2['Year Acquired'] != 0]
    df2 = df2.groupby('Year Acquired')['Square Footage'].sum()
    return df2

mySelectedLabel = ipywidgets.Label()
def on_selected(change):
    if len(change['owner'].selected) == 1: # only allow user to select 1 bin
        i,j = change['owner'].selected[0]
        v = mean_sq_ft.values[i,j] # data value in that bin
        mySelectedLabel.value = 'Mean Square Footage for Agency: ' + mean_sq_ft.index[i] + ', and County: ' + mean_sq_ft.columns[j] + ' = ' + str(v)
        buildings_final2 = barplot_fig2(mean_sq_ft.index[i], mean_sq_ft.columns[j])
        #print(buildings_final2.values)
        bar_plot.x = mean_sq_ft.index
        bar_plot.y = mean_sq_ft.values
        #mySelectedLabel.value = plt.bar(buildings_final2.index, buildings_final2.values)
heat_map.observe(on_selected, 'selected')

# 6. figure!
fig = bqplot.Figure(marks=[heat_map], axes=[col_ax, x_ax, y_ax], fig_margin={'top':30, 'bottom':60, 'left':80, 'right':100})

# fig.layout.min_width='500px'
# fig_dur.layout.min_width = '500px'
# figures = ipywidgets.HBox([fig, fig_dur])

myDashboard = ipywidgets.VBox([mySelectedLabel, fig, fig_dur])
myDashboard

# myDashboard = ipywidgets.VBox([fig, fig_dur])
# myDashboard
```

#### The above function. is created to display the heatmap. The scale is ordinal as the z-axis and y-axis is dicrete. I have used the heatmap from bqplot and passed the values of Square footage as color to display the values of mean square footage by 'Agency Name' and 'County'. The min and max values are specified and a comment for user readibility is included since the names are overlapping while displaying. For the gridmap, the rows and columns were assigned according to the pivot table. The selected value would be highlighted in blue color. Also, selected the appropriate marks, scales and axis to finally display the figure. 

#### The function created is for an onclick event named on selected and it displays the mean values and get updates every time the selection changes, Next, the function barplot_fig2 takes the user selected values and retrives the same values from our dataset. Now, the values of year acquired that are equal to zero are dropped. The sum of the year are stored into a dataframe. Then a callback is registered using Observe method so when the user changes the value, observe method returns a call back to the on_selected function.

#### The heatmap is plotted based on the axes and marks specified along with the customizations. Some selections are made to increase user readability like adding the comment above the gridmap and not displaying the figured horizontally. The axes overlaps each other. The figures are linked using on_selected function and plotted suing VBox function.
